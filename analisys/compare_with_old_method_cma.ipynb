```python
def shapiro_wilk(df):
    try:
        stat, p_value = stats.shapiro(df['P_B'])
        return {'Estatística de Shapiro-Wilk': stat, 'Valor p': p_value}
    except Exception as e:
        print('Erro ' + str(e))
        return {'Erro': str(e)}
```

```python

def run_statistical_test(df_a, df_b, alpha=0.05):
    
    df_a_result = shapiro_wilk(df_a)
    df_b_result = shapiro_wilk(df_b)
    
    print('Shapiro A', df_a_result['Valor p'])
    print('Shapiro B', df_b_result['Valor p'])
    
    # If p-values less than 0.05 (Significance Level) use Mann-Whitney U, otherwise, use T-Test
    if df_a_result['Valor p'] > alpha and df_b_result['Valor p'] > alpha:
        # Use T-Test
        t_statistic, p_value = stats.ttest_ind(df_a['P_B'], df_b['P_B'])
        return {'Algorithm':'t-test', 'Statistic':t_statistic, 'p-value':p_value, 'difference': (p_value < alpha)}
    else:
        
        # Use Mann-Whitney U
        statistic, p_value = stats.mannwhitneyu(df_a['P_B'], df_b['P_B'], alternative='two-sided')
        return {'Algorithm':'mann-whitney-u', 'Statistic':statistic, 'p-value':p_value, 'difference': (p_value < alpha)}
        
```

```python
alpha = 0.05

result = run_statistical_test(old_df_7, new_limitless_df_7)

print('Algorithm:', result['Algorithm'])
print('Statistic:', str(result['Statistic']))
print('P-value:', str(result['p-value']))
print('Difference?:', str(result['difference']))
print()
```
